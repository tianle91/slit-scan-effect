```python
import numpy as np
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture('source/IMG_3643.MOV')

width, height, fps = int(cap.get(3)), int(cap.get(4)), cap.get(5)

def get_vertical_slice(frame):
    center = int(width/2)
    return frame[:, (center-5):(center+5), :]

ret = True
i = 0
vertical_slices = []

while ret:
    print ('frame: {}'.format(i))
    ret, frame = cap.read()
    i += 1
    if frame is not None:
        vertical_slices.append(get_vertical_slice(frame))
        
cap.release()
```

```python
img = np.concatenate(vertical_slices[::-1], axis=1)
```

```python
img.shape
```

```python
plt.figure(figsize=(14*(img.shape[1]/img.shape[0]), 4))
plt.imshow(img)
```

```python
cv2.imwrite('output.jpg', img)
```

```python
# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, fps, (width, height))

for start_horizontal in range(0, img.shape[1]-width, int((img.shape[1]-width)/i)):
    print (start_horizontal)
    frame_out = img[:, start_horizontal:(start_horizontal+width), :]
    out.write(frame_out)
    
out.release()
```
